# **priv 데이터 셋 구성**

```python
df_priv_info = pd.read_csv('/content/drive/Shareddrives/데캡디/Result/Recommendation_data/raw_x_y_priv.csv')
df_priv_info.head()
```

```python
df_priv_v1 = df_priv.copy()
df_priv_v1 = df_priv_v1[['JOIN_SN', 'result', 'LN_AMT_max', 'LN_AMT_min', 'RATE_max', 'RATE_min']]

tmp_result = df_priv_v1['result'].str.split(',')
tmp_result = tmp_result.apply(lambda x: pd.Series(x))

tmp_result.columns = ['LN_CD_1', 'LN_CD_2', 'LN_CD_3']

df_priv_v1 = pd.concat([df_priv_v1, tmp_result], axis = 1)
df_priv_v1 = pd.merge(df_priv_v1, df_priv_info, how = 'left', left_on = 'JOIN_SN', right_on = 'JOIN_SN')
df_priv_v1 = df_priv_v1.drop(['result', 'is_applied'], axis = 1)


df_priv_v1 = df_priv_v1.astype({'LN_CD_1' : 'float64',
                                'LN_CD_2' : 'float64',
                                'LN_CD_3' : 'float64'})

df_priv_v1 = df_priv_v1.replace({'LN_CD_1' : 31}, '개인 대출')
df_priv_v1 = df_priv_v1.replace({'LN_CD_1' : 41}, '장기 카드 대출 (카드론)')
df_priv_v1 = df_priv_v1.replace({'LN_CD_1' : 37}, '단기 카드 대출 (현금서비스)')

df_priv_v1 = df_priv_v1.replace({'LN_CD_2' : 0}, '카드 대출')
df_priv_v1 = df_priv_v1.replace({'LN_CD_2' : 100}, '신용 대출 > 신용 대출')
df_priv_v1 = df_priv_v1.replace({'LN_CD_2' : 150}, '신용 대출 > 학자금 대출')

df_priv_v1 = df_priv_v1.replace({'LN_CD_3' : 0}, '서민 금융 아님')

df_priv_v1 = df_priv_v1.replace({'GENDER' : 1}, '남성')
df_priv_v1 = df_priv_v1.replace({'GENDER' : 2}, '여성')

df_priv_v1 = df_priv_v1.replace({'CD_USG_LMT' : 0}, '-')
df_priv_v1 = df_priv_v1.replace({'CD_CA_LMT' : 0}, '-')
df_priv_v1 = df_priv_v1.replace({'CD_USG_AMT' : 0}, '-')
df_priv_v1 = df_priv_v1.replace({'CD_CA_AMT' : 0}, '-')
df_priv_v1 = df_priv_v1.replace({'GRNT_AMT' : 0}, '-')


df_priv_v1 = df_priv_v1.fillna('-')

# df_priv_v1 = df_priv_v1.rename(columns = {
#     'JOIN_SN' : '고객 번호'
#     ''


# })


df_priv_v1
```
