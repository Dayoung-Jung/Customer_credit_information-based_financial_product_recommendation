# **데이터 합치기**

```python
df_priv_y = pd.read_csv('/content/drive/Shareddrives/데캡디/Result/Raw_data/y_final_priv.csv')
df_priv_y
```

```python
df_priv_v2 = pd.merge(df_A_ID, df_A_DLQ_PUB, how='left', left_on='JOIN_SN', right_on='JOIN_SN')
df_priv_v2 = pd.merge(df_priv_v2, df_P_CD_OPN, how='left', left_on='JOIN_SN', right_on='JOIN_SN')
df_priv_v2 = pd.merge(df_priv_v2, df_P_CD_USG, how='left', left_on='JOIN_SN' , right_on='JOIN_SN')
df_priv_v2 = pd.merge(df_priv_v2, df_P_GRNT, how='left', left_on = 'JOIN_SN' , right_on='JOIN_SN')
df_priv_v2 = pd.merge(df_priv_v2, df_priv_y, how='left', left_on='JOIN_SN' , right_on='JOIN_SN')

df_priv_v2 = df_priv_v2.drop_duplicates()

df_priv_v2
```

```python
df_priv_v2.isnull().sum()
```

```python
df_priv_v2.columns
```

# **카드 데이터만 추출 + 금융상품**

```python
df_priv_v2 = df_priv_v2[['JOIN_SN', 'BTH_YR', 'GENDER', 'CD_OPN_SCTR_CD', 'CD_OPN_COM_SN', 
                         'CD_OPN_CD_1', 'CD_OPN_CD_2', 'CD_USG_SCTR_CD', 'CD_USG_COM_SN', 
                         'CD_USG_LMT', 'CD_CA_LMT', 'CD_USG_AMT', 'CD_CA_AMT', 'is_applied']]

df_priv_v2['is_applied'].fillna(0,inplace=True)
df_priv_v2 = df_priv_v2.dropna(axis=0)
df_priv_v2 = df_priv_v2.drop_duplicates()

df_priv_v2
```

```python
# 개인 대출 정보 (금융업권)
df_P_LN = pd.read_csv(path + 'SDB2_P_LN.csv')
df_P_LN = df_P_LN.copy()

df_P_LN = df_P_LN[['JOIN_SN', 'LN_CD_1', 'LN_CD_2', 'LN_CD_3', 'LN_DT', 'LN_AMT', 'RATE']]
df_P_LN = df_P_LN.drop_duplicates()

df_P_LN
```

```python
# 개인 대출 정보 (대부업권)

df_P_LN_SP = pd.read_csv(path + 'SDB2_P_LN_SP.csv')
df_P_LN_SP = df_P_LN_SP.copy()

df_P_LN_SP = df_P_LN_SP[['JOIN_SN', 'LN_CD_1', 'LN_CD_2', 'LN_CD_3', 'LN_DT', 'LN_AMT', 'RATE']]
df_P_LN_SP = df_P_LN_SP.drop_duplicates()

df_P_LN_SP
```

```python
concat_df = pd.concat([df_P_LN, df_P_LN_SP], ignore_index = True)
concat_df = concat_df.groupby(['JOIN_SN',	'LN_CD_1', 'LN_CD_2', 'LN_CD_3',	'LN_DT']).mean().reset_index()
concat_df
```

```python
df_final_priv = pd.merge(df_priv_v2, concat_df, how='left', left_on='JOIN_SN', right_on='JOIN_SN')

df_final_priv = df_final_priv.drop_duplicates()
df_final_priv
```
