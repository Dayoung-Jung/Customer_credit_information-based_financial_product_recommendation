#### 금리, 대출 금액 정리 데이터

```python
index_list = df_selected.reset_index()['index'].to_list()
```

```python
df_selected
```

```python
j = 0
df_result = []
temp_list_max = []
temp_list_min = []
temp_listR_max = []
temp_listR_min = []

while j != 4135:
    df_serial = list(zip(df_priv_raw_cs[j], df_priv_raw_cs['JOIN_SN']))
    sim_scores = sorted(df_serial, key = lambda x: x[0], reverse = True)
    sim_scores = sim_scores[1:51]
    sim_scores
    df_indices = [idx[1] for idx in sim_scores]
    df_indices
    q = index_list[j]
    temp1 = df_selected['result'][q].replace(' ', '').split(',')[0]
    temp2 = df_selected['result'][q].replace(' ', '').split(',')[1]
    temp3 = df_selected['result'][q].replace(' ', '').split(',')[2]
    temp1 = float(temp1)
    temp2 = float(temp2)
    temp3 = float(temp3)
    temp = df_priv.query('JOIN_SN in @df_indices')
    temp = temp.query('LN_CD_1 == @temp1')
    temp = temp.query('LN_CD_2 == @temp2')
    temp = temp.query('LN_CD_3 == @temp3')
    temp_max = temp['LN_AMT'].max()
    temp_min = temp['LN_AMT'].min()
    tempR_max = temp['RATE'].max()
    tempR_min = temp['RATE'].min()
    temp_list_max.append(temp_max)
    temp_list_min.append(temp_min)
    temp_listR_max.append(tempR_max)
    temp_listR_min.append(tempR_min)
    j+=1

df_selected['LN_AMT_max'] = temp_list_max
df_selected['LN_AMT_min'] = temp_list_min
df_selected['RATE_max'] = temp_listR_max
df_selected['RATE_min'] = temp_listR_min
df_selected
```
