## **연체 정보 (신용도 판단 정보)**

```python
# 데이터 불러오기
df_A_DLQ = pd.read_csv(path + 'SDB2_A_DLQ.csv')
df_A_DLQ = df_A_DLQ.copy()


# 기본 셋팅
df_A_DLQ = df_A_DLQ[df_A_DLQ['JOIN_SN_TYP'] == 1]
df_A_DLQ = df_A_DLQ.drop(columns = ['YM', 'JOIN_SN_TYP', 'IS_ME'], axis = 1)


# 연제 등록 월: 카운트
# # 연체 등록 금액: 평균, 최솟값, 최댓값
df_A_DLQ_RE = df_A_DLQ.groupby(["JOIN_SN"]).agg({"DLQ_RGST_AMT":["mean", "min", "max"], "DLQ_RGST_DT" : pd.Series.nunique}).reset_index()
df_A_DLQ_RE.columns = ['JOIN_SN', 'DLQ_RGST_AMT_mean', 'DLQ_RGST_AMT_min', 'DLQ_RGST_AMT_max', 'DLQ_RGST_DT_count']


# 연체 사유 코드 1 경험 유무
# 유: 1 / 무: 0
df_A_DLQ_CD = df_A_DLQ[['JOIN_SN', 'DLQ_CD_1', 'DLQ_CD_2', 'DLQ_RGST_DT']]
df_A_DLQ_CD = df_A_DLQ_CD.drop_duplicates()

DLQ_CD_1_list = [101, 201, 202, 103, 104, 115, 112, 105, 109, 203, 102, 116, 208, 401, 113, 207, 403, 402, 117]

for i in DLQ_CD_1_list:
  df_A_DLQ_CD.loc[df_A_DLQ_CD['DLQ_CD_1'] == i,'DLQ_CD_1_' + str(i)] = 1


DLQ_CD_2_list = [0, 7]

for i in DLQ_CD_2_list:
  df_A_DLQ_CD.loc[df_A_DLQ_CD['DLQ_CD_2'] == i,'DLQ_CD_2_' + str(i)] = 1


df_A_DLQ_CD.fillna(0,inplace=True)
df_A_DLQ_CD = df_A_DLQ_CD.drop(columns = ['DLQ_CD_1', 'DLQ_CD_2', 'DLQ_RGST_DT'], axis=1)
df_A_DLQ_CD = df_A_DLQ_CD.groupby('JOIN_SN').sum().reset_index()



# 업권 코드 경험 유무
# 유: 1 / 무: 0
df_A_SCTR_CD = df_A_DLQ[['JOIN_SN', 'SCTR_CD']]
SCTR_CD_list = [1, 2, 3, 5, 6, 8, 10, 12, 13, 15, 16, 17, 21, 21]

for i in SCTR_CD_list:
  df_A_SCTR_CD.loc[df_A_SCTR_CD['SCTR_CD'] == i,'SCTR_CD_' + str(i)] = 1
  
df_A_SCTR_CD.fillna(0,inplace=True)
df_A_SCTR_CD = df_A_SCTR_CD.drop(columns = ['SCTR_CD'], axis=1)
df_A_SCTR_CD = df_A_SCTR_CD.drop_duplicates()
df_A_SCTR_CD = df_A_SCTR_CD.groupby('JOIN_SN').sum().reset_index()
```

## **연체 정보 (공공 정보)**

```python
# 데이터 불러오기
df_A_PUB = pd.read_csv(path + 'SDB2_A_PUB.csv')
df_A_PUB = df_A_PUB.copy()

# 기본 셋팅
df_A_PUB = df_A_PUB[df_A_PUB['JOIN_SN_TYP'] == 1]
df_A_PUB = df_A_PUB.drop(columns = ['JOIN_SN_TYP', 'IS_ME'], axis = 1)

# 연제 등록 월: 카운트
# 연체 등록 금액: 평균, 최솟값, 최댓값
df_A_PUB_RE = df_A_PUB.groupby(["JOIN_SN"]).agg({"DLQ_RGST_AMT":["mean", "min", "max"], "DLQ_RGST_DT" : pd.Series.nunique}).reset_index()
df_A_PUB_RE.columns = ['JOIN_SN', 'PUB_DLQ_RGST_AMT_mean', 'PUB_DLQ_RGST_AMT_min', 'PUB_DLQ_RGST_AMT_max', 'PUB_DLQ_RGST_DT_count']



# 연체 사유 코드 1 경험 카운트
df_A_PUB_DLQ_CD = df_A_PUB[['JOIN_SN', 'DLQ_CD_1', 'DLQ_RGST_DT']]
df_A_PUB_DLQ_CD = df_A_PUB_DLQ_CD.drop_duplicates()

PUB_DLQ_CD_1_list = [601, 1201,  801, 1101,  604, 1301,  802,  606,  611,  605,  602, 807,  803, 1701]

for i in PUB_DLQ_CD_1_list:
  df_A_PUB_DLQ_CD.loc[df_A_PUB_DLQ_CD['DLQ_CD_1'] == i,'PUB_DLQ_CD_1_' + str(i)] = 1

df_A_PUB_DLQ_CD.fillna(0,inplace=True)
df_A_PUB_DLQ_CD = df_A_PUB_DLQ_CD.drop(columns = ['DLQ_CD_1', 'DLQ_RGST_DT'], axis=1)
df_A_PUB_DLQ_CD = df_A_PUB_DLQ_CD.groupby('JOIN_SN').sum().reset_index()


# 업권 코드 경험 유무
# 유: 1 / 무: 0
df_A_SCTR_CD = pd.merge(df_A_SCTR_CD, df_A_PUB[['JOIN_SN', 'SCTR_CD']], how='outer', left_on='JOIN_SN', right_on='JOIN_SN')
df_A_SCTR_CD.loc[df_A_SCTR_CD['SCTR_CD'] == 12,'SCTR_CD_12'] = 1
df_A_SCTR_CD.fillna(0,inplace=True)
df_A_SCTR_CD = df_A_SCTR_CD.drop(columns = ['SCTR_CD'], axis=1)
df_A_SCTR_CD = df_A_SCTR_CD.drop_duplicates()
df_A_SCTR_CD = df_A_SCTR_CD.groupby('JOIN_SN').sum().reset_index()
```

## **신용카드 이용정보**

```python
# 데이터 불러오기
df_P_CD_USG = pd.read_csv(path + 'SDB2_P_CD_USG.csv')
df_P_CD_USG = df_P_CD_USG.copy()

# 기본 셋팅
df_P_CD_USG = df_P_CD_USG[df_P_CD_USG['JOIN_SN_TYP'] == 1]
df_P_CD_USG = df_P_CD_USG.drop(columns = ['YM', 'COM_SN', 'IS_ME'], axis = 1)

# 신용판매한도금액: 평균값
# 현금서비스한도금액: 평균값
# 신용판매이용금액: 합계
# 현금서비스이용금액: 합계
df_P_CD_USG = df_P_CD_USG.groupby(["JOIN_SN"]).agg({"CD_USG_LMT":"mean", "CD_CA_LMT":"mean", "CD_USG_AMT":"sum", "CD_CA_AMT":"sum"}).reset_index()
```
